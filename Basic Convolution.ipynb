```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

#### [Image Kernels](http://setosa.io/ev/image-kernels/)



```python
img = cv2.imread("/Users/Pratulya/Desktop/attendance1.png")
print(img.shape)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, (500,500))
print(img.shape)
```

```python
plt.figure()
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# kernel = np.ones((3,3))/9 # Blur

# kernel = np.array([ # Gaussian Blur
#     [0.0625, 0.125, 0.0625],
#     [0.125,  0.25,  0.125],
#     [0.0625, 0.125, 0.0625],
# ])

kernel = np.array([
    [1,0,-1],
    [2,0,-2],
    [1,0,-1],
])
kernel.shape, kernel.sum()
```

```python
convolved_img = np.zeros((img.shape[0] - kernel.shape[0] + 1, img.shape[1] - kernel.shape[1] + 1))
convolved_img.shape
```

```python
for i in range(convolved_img.shape[0]):
    for j in range(convolved_img.shape[1]):
        patch = img[i:i+kernel.shape[0], j:j+kernel.shape[1]]
        convolved_img[i,j] = (patch * kernel).sum()
```

```python
plt.figure()
plt.imshow(img, cmap='gray')
plt.figure()
plt.imshow(convolved_img, cmap='gray')
plt.show()
```
